```python
import pandas as pd
```

```python
data = pd.read_csv('all_concepts.csv', delimiter=":", header=None)
```

```python
data.head()
```

```python
res = []
for index, row in data.iterrows():
    count = 0 
    for i in row[1].split(';'):
        count = count + 1 
    temp = str(row[0])+';'+str(count)
    res.append(temp)
```

```python
final_results = pd.DataFrame(
    {
        "0":res,
    })
print final_results
```

```python
final_results.to_csv('classfeg.txt',  index = False, header=False)
```
